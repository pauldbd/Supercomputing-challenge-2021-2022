```python
f['state'].replace(['control', 'mild cognitive impairment', "Alzheimer's disease"],
                        [0, 1,2], inplace=True)
f.head()
```

```python
final = f.copy(deep = True)
final.head()
```

```python
#virtually no constant features 
X = final.copy()
X = X.iloc[: , :-1]
y = X["state"]
X.drop(labels=['state'], axis=1, inplace = True)
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=0)
sel.fit(X)
sum(sel.get_support())
```

```python
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
fvalue_Best = SelectKBest(f_classif, k=10)
X_kbest = fvalue_Best.fit_transform(X, y)
print(X_kbest)
cols = fvalue_Best.get_support(indices=True)
features_df_new = X.iloc[:,cols]

print('Original numbermm of features:', X.shape)
print('Reduced number of features:', X_kbest.shape)
```

```python
fit = fvalue_Best.fit(features_df_new,y)
print(fit.scores_)
print(fit.pvalues_)
```
