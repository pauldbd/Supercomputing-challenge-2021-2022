```python
#virtually no constant features 
X = final.copy()
y = X["state"]
X.drop(labels=['state'], axis=1, inplace = True)
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=0)
sel.fit(X)
sum(sel.get_support())
```

```python
#information gain 
from sklearn.feature_selection import mutual_info_classif as MIC
mi_score = list(MIC(X,y))
```

```python
import numpy as np
arr = np.array(mi_score)
```

```python
mi_score_ind = np.where(arr > 0.18)[0]
```

```python
print(len(mi_score_ind))
```

```python
#this is the most strongly associated with ad
print(X.columns[mi_score_ind])
```

```python
#make it less restrictive for pearson corr
mi_score_ind = np.where(arr > 0.07)[0]
print(len(mi_score_ind))
```

```python
print(mi_score_ind)
```

```python
new = pd.DataFrame()
for i in mi_score_ind:
    new.loc[:, X.columns[i]] = X.iloc[:, i]


    
```

```python
new.insert(0,"state",y,allow_duplicates = True)
new.head()
```

```python
new.columns[0]
```

```python
from pandas import *
import numpy as np
from scipy import stats
import itertools
correlations = {}
bestPos = []
bestNeg = []
alls = []
columns = new.columns.tolist()
#Find out which methylated regions are most correlated with ad using pearsonr correlation. For the machine learning model, the filters are less restrictive. 
for col_a, col_b in itertools.combinations(columns, 2):
    if col_a == 'state':
        r,p = stats.pointbiserialr(new.loc[:, col_a], new.loc[:, col_b])
        correlations[col_a + '__' + col_b] = stats.pointbiserialr(new.loc[:, col_a], new.loc[:, col_b])
        #differentiate between positively correlated with ad and negatively correlated - neg or pos pearsonr val
        if r > 0.25 and p < 0.05:
            bestPos.append(str(col_b))
            alls.append(str(col_b))
        elif r < -0.25 and p <0.05:
            bestNeg.append(str(col_b))
            alls.append(str(col_b))

result = DataFrame.from_dict(correlations, orient='index')
result.columns = ['PCC', 'p-value']
print(result.sort_index(ascending=False))
```

```python
print(len(bestPos))
print(len(bestNeg))
print(bestPos)
print("")
print("")
print("")
print("")
print(bestNeg)
```

```python
theNew = pd.DataFrame()
for (columnName, columnData) in new.iteritems():
    if columnName in alls:
        theNew[columnName] = columnData
```
